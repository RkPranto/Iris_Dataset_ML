```python
%matplotlib inline
import seaborn as sb
iris = sb.load_dataset('iris')
correlation = iris.corr()
correlation
```

```python
sb.heatmap(correlation, cmap='YlGnBu', annot=True)
```

```python
sb.set(style="whitegrid", color_codes=True)
sb.lmplot(x="sepal_length", y = "petal_length", hue = "species", data = iris, palette="Set1", markers=["o", "s", "D"])
```

```python
 sb.pairplot(iris, height=5, vars=["sepal_length", "petal_length"], markers=["o","s", "D"], hue="species")
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
print(knn)
```

```python
knn.fit(iris.data, iris.target)
```

```python
knn.predict([[3,5,4,2]])
```

```python
print("Predicted target name: ", iris['target_names'][knn.predict([[3,5,4,2]])])
```
